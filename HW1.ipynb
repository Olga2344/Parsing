```python
params={
    'latitude':'55.7577374',
    'longitude':'37.6164793',
    'query': 'пепси'
}

delivery_respons = requests.get('https://api.delivery-club.ru/api1.2/vendors/search', params=params)
```

```python
vendor_names=delivery_respons.json()['vendors']
```

```python
pprint(delivery_respons.json()['vendors'][0])
```

Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search

(а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. Для каждой категории рассчитать среднюю минимальную стоимость заказа.
(б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

```python
vendor_free_delivery=[]
vendor_payment_dlvr=[]
cost_of_dlvr_lst=[]

for vendor in vendor_names:

    if vendor['delivery']['price']['value']>1:
        vendor_payment_dlvr.append(vendor['name'])
        cost_of_dlvr_lst.append(vendor['delivery']['price']['value'])
    else:
        vendor_free_delivery.append(vendor['name'])
                  
print( 'Бесплатная доставка: \n\n', vendor_free_delivery, '\n\nПлатная доставка: \n\n',vendor_payment_dlvr)
print('\n\nСредняя стоимтость платной доставки: \n\n',np.mean(cost_of_dlvr_lst))
```

```python
x=(len(vendor_free_delivery)*100)/(len(vendor_free_delivery)+len(vendor_payment_dlvr))
print(f'% платной доставки: {int(x)}%,\n\n% бесплатной доставки: {100-int(x)}%')
```

```python
free_delivery=0
vendor_min_order_1=[]
vendor_whith_limith=[]
cost_of_dlvr_lst_2=[]

for vendor in vendor_names:

    if vendor['delivery']['minOrderPrice']['value']>1:
        vendor_whith_limith.append(vendor['name'])
        cost_of_dlvr_lst_2.append(vendor['delivery']['minOrderPrice']['value'])
    else:
        vendor_min_order_1.append(vendor['name'])
                  
print( 'Минимальный заказ 1 рубль: \n\n', vendor_min_order_1, '\n\nМинимальный заказ больше 1 рубля: \n\n',vendor_whith_limith)
print('\n\nсредняя стоимтость минимального зказа: \n\n',np.mean(cost_of_dlvr_lst_2))
```

```python
cafe_vendor=[]
shop_vendor=[]
for vendor in vendor_names:
    if vendor['categoryId']==1:
        cafe_vendor.append(vendor['name'])
    else:    
        shop_vendor.append(vendor['name'])
        
print('Кафе: \n\n', cafe_vendor, '\n\nМагазины и ларьки: \n\n',shop_vendor)
```

```python
x=(len(cafe_vendor)*100)/(len(cafe_vendor)+len(shop_vendor))
print(f'Кафе: {int(x)}%,\n\nМагазины и ларьки: {100-int(x)}%')
```
