```python
params={
    'latitude':'55.7577374',
    'longitude':'37.6164793',
    'query': ''
}

delivery_respons = requests.get('https://api.delivery-club.ru/api1.2/vendors/search', params=params)
```

```python
vendor_names=delivery_respons.json()['vendors']
```

```python
pprint(delivery_respons.json()['vendors'][0])
```

     API.       https://www.delivery-club.ru/search

```python
vendor_free_delivery=[]
vendor_payment_dlvr=[]
cost_of_dlvr_lst=[]

for vendor in vendor_names:

    if vendor['delivery']['price']['value']>1:
        vendor_payment_dlvr.append(vendor['name'])
        cost_of_dlvr_lst.append(vendor['delivery']['price']['value'])
    else:
        vendor_free_delivery.append(vendor['name'])
                  
print( ' : \n\n', vendor_free_delivery, '\n\n : \n\n',vendor_payment_dlvr)
print('\n\n   : \n\n',np.mean(cost_of_dlvr_lst))
```

```python
x=(len(vendor_free_delivery)*100)/(len(vendor_free_delivery)+len(vendor_payment_dlvr))
print(f'%  : {int(x)}%,\n\n%  : {100-int(x)}%')
```

```python
free_delivery=0
vendor_min_order_1=[]
vendor_whith_limith=[]
cost_of_dlvr_lst_2=[]

for vendor in vendor_names:

    if vendor['delivery']['minOrderPrice']['value']>1:
        vendor_whith_limith.append(vendor['name'])
        cost_of_dlvr_lst_2.append(vendor['delivery']['minOrderPrice']['value'])
    else:
        vendor_min_order_1.append(vendor['name'])
                  
print( '  1 : \n\n', vendor_min_order_1, '\n\n   1 : \n\n',vendor_whith_limith)
print('\n\n   : \n\n',np.mean(cost_of_dlvr_lst_2))
```

```python
cafe_vendor=[]
shop_vendor=[]
for vendor in vendor_names:
    if vendor['categoryId']==1:
        cafe_vendor.append(vendor['name'])
    else:    
        shop_vendor.append(vendor['name'])
        
print(': \n\n', cafe_vendor, '\n\n  : \n\n',shop_vendor)
```

```python
x=(len(cafe_vendor)*100)/(len(cafe_vendor)+len(shop_vendor))
print(f': {int(x)}%,\n\n  : {100-int(x)}%')
```
